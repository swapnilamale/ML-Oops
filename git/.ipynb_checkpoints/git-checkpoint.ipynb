```
# Get username & email
    git config user.name
    git config user.email

# Set username & email
    git config --global user.name <username>  
    git config --global user.email <email>

# Set up Default editor as VS code for git commit messages (Default is VIM)
    git config --global core.editor "code --wait"
```

### Initialising & Commiting Changes to Git

```
# Inititalising git commit
    git init      : Create pwd as git repo
                  : it create .git hidden folder in the pwd which will track the changes
                  : it is one time command only use when to make pwd as git repo
                  : Itâ€™s generally a bad practice to initialize a Git repository inside another Git repository because 
                    it creates nested .git folders, which can lead to confusing version control behavior 
                    and difficult debugging.
              
-------------------------------------------------------------------------------------------------------------------- 

                    Working Directory --------------> Staging Area -------------------> Repository
                                        git status                      git commit
                                         git add                         git log                   
                                        
# Check Changes to commit in stagging
    git status


# Adding changes from working directory to staging for group commiting
    git add <file1> <file2>
    git add .  
    git add -A    # - Sometimes, depending on your Git version or configuration, 
                      running "git add ." may not stage deleted files automatically.
                      it only track "new files" and "modified files"
                    - This can lead to a situation where deleted files are still shown in git status 
                      and not removed in the next commit.
                    - To avoid this inconsistency, it's a good practice to always use -A




# Adding changes from staging to repository
    git commit                   # Commiting with default editior for message writing (VIM)
    git commit -m "message"      # Quick commit  
    git commit --amend           # Commiting changes to the PREVIOUS commit 
                                   - fixing previous commit message
                                   - adding new files in the staging to previous commit
                               
Note: Automatic Commit (Keep each commit focused on a single thing)
      Commited Messages should be in present tense
      

# Git Log : return log of the commits for the given repository (Author,Timestamp,Commit_Message)
    git log                      
    git log --decorate
    git log --oneline 
    git log --oneline --decorate
```

### .gitignore file

```
- .gitignore file tells Git which files and folders it should ignore and not track.
- Git does not track empty folders by default.
  So even if you list certain files inside a folder in .gitignore, 
  the folder itself will disappear from your repo if it   has no tracked files.
  If you still want Git to keep the folder structure, you must add a placeholder file inside the folder
```

```
# list of git branches
    git branch  
    git branch -v 
    

# Creating branch
    git branch <branch_name>            



# switching head to new branch
    git switch <existing_branch_name>   git chekout <existing_branch_name>   : switching to another existing branch
    git switch -c <new_branch>          git checkout -b <new_branch>         : creating and switching to new_branch
       (new version of git)                  (older verseron of git)

Note: Always Commit or Stash Changes Before Switching Branches
      Before switching branches in Git, you should add and commit your changes (or stash them).
      If you don't, your uncommitted work may: - Get overwritten
                                               - Cause merge conflicts
                                               - lost depending on the branch state



# Rename Current Branch (First swtich to branch need to rename)
    git branch -M <new_name>    : -m is for Move/Rename
                                  -M for force Move/Rename

# Deleting branch (First switch to another branch)
    git branch -D <branch_name>   : -d for delete
                                    -D for force delete
```

- You want to switch branches but you have uncommitted changes, git wonâ€™t allow switching because of conflict risk 
  else git will take thoes changes to other branch as unstaged changes
  in that case stashing is used to Temporarily save your work without making unnecessary commits
- git stash is super useful command that helps you save changes that you are not yet ready to commit
```

```
git stash | git stash save   : will take all uncommited changes (staged and unstaged) and stash them 
                               and remove changes in working directory
git stash pop                : remove the most recently shashed changes in your stash 
                               and add changes to working directory
git stash apply              : add changes to working directory 
                               without removing from the stash
                               
                               
git stash list               : list of git stash available to pop or apply in ORDER
git stash apply <stash_id>   : applying changes from the perticular stash
git stash drop  <stash_id>   : deleting stash from the stash list
git stash clear              : clearing everything from git stash list
```

<div style="text-align: center; color: blue;">
    <h1>Chapter 6 : Undoing Changes & Time Traveling</h1>
</div>

```

There are two types of branches in git
    1. Local Branches            (git branch)
    2. Remote-Tracking Branches  (git branch -r)
    
remote-tracking branches?
- Remote-tracking branches are read-only local references to the state of branches on a remote repository.
- A remote-tracking branch is a local reference that tracks the state of a branch on a remote repository.
- Once you pull git remote code to local, git by defualt create one read-only branch having name 
  <remote_name>/<branch_name> just to know how the remote looked like the last time I checked
- remote-tracking branches are NOT a real local branch, just a pointer

        git branch -r : Shows all remote-tracking branches (<remote_name>/<branch_name>)
                        Ex: origin/main
                            origin/develop
                            origin/feature-x
                            
- Once you have made multiple commits locally and want to see how the code looked 
  when you initially cloned the repository, you can check out the remote-tracking branch (origin/main).
                                  git ckeckout origin/main
  Note: This puts the repository in a detached HEAD state and should be used only for viewing, not for making changes.
  
  
- Once you push your code to the remote repository, the origin/main pointer is updated to the latest pushed commit.
```

### git fetch

```
git fetch  : - command fetches branches and history from a specific remote repository
               and it only updates remote-tracking branches
             - Fetching allows us to download changes from a remote repository, BUT
               those changes will not be automatically integrated into our working files
             - it lets you see what others have been working on,
               without having to merge those changes into your local repo
             - Its just like "please go and get the latest information from Github, BUT 
               don't mess up my working directory"
             
             
             git fetch <remote_name>
             git fetch origin
```

### git pull

```
git pull : - command we can use to retrieve changes from a remote repository.
             Unlike fetch, pull actually updates local-branches along with remote-tracking branches
             with whatever changes are retrieved from the remote
           - Its just like " download data from Github AND immediately update my local repo with those changes"
           - git pull   =       git fetch         +       git merge
                          update remote-trackting    update local branches
                                 branches
           
           
           git pull <remote_name> <branch_name>  : It will pull all the changes in <branch_name> on remote
                                                   and merge thoes changes with CURRENT BRANCH
                                                   Hence git pull CAN create merge conflicts
                                                   
```

<div style="text-align: center; color: blue;">
    <h1>Chapter 7: Github Grab Bag: Odds & Ends</h1>
</div>

### github repo visibility : Public Vs. Private

```
