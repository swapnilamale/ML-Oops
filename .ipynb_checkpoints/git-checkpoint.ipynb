```
# What is Git?
- Worlds most popular Free and Open Source VCS (Version Control System)
- It is use to SAVE different checkpoints of code, CREAT new branch from aribitory checkpoints, COMBINE checkpoints

# use of Git
- Track changes across multiple files
- Compare versions of a project
- "Time Travel" back to old versions
- Revert to a previous version
- Collaborate and share changes : Specific usecase of github
- Combine changes

# History of Git
Linus Torvalds a creator of git and linux
2005 : Release first version of git with basic functionality


# Who Uses Git
Engineers, Coader, Researcher, Organisations, Governments, Writers


# Git vs GitHub
            Git                                                 GitHub
- its a software which install locally (VCS)    - Github is a service that hosts GitRepo in cloud & 
                                                  makes it easier to collaborate with other people
- No internet needed                            - internet needed



# Interaction with Git

             Terminal                                     GUI
        - Command base                   - Github Desktop/ SourceTree/ Tower/ GitKraken/ Ungit
        - More Faster                    - Good for beginers 
```

```
What is git repo?
Git Repo : A workspace folder which tracks changes in files within folder
           Every Git Repo has its own history which save's checkpoints of code called git commits
           folder labeled as git repo when it contains ".git" file which is created by "git init" command
           if we delete .git folder, repo become normal local directory
```

### Configiration Commands

```
# Get username & email
git config user.name
git config user.email

# Set username & email
git config --global user.name <username>  
git config --global user.email <email>

# Set up Default editor as VS code for git (Default is VIM)
git config --global core.editor "code --wait"
```

### Initialising & Commiting Changes to Git

```
# Inititalising git commit
git init      : Create pwd as git repo
              : it create .git hidden folder in the pwd which will track the changes
              
              
              
# Adding changes from working directory to staging for group commiting
git add <file1> <file2>
git add .  



# Check Changet to commit in stagging
git status



# Adding changes from staging to repository
git commit -m "message"      # Quick commit  
git commit                   # Commiting with default editior
git commit --amend           # Commiting changes to the previos commit
git commit -a -m "message"   # adding all changes to staging and commiting with message in single command

Note: Automatic Commit
      Commit Message should be in present tense
 
 

# Git Log
git log                 : return log of the commits for the given repository (Author,Timestamp,Commit_Message)
git log --oneline
```

    <h1>Chapter 3 : Git Branching</h1>
</div>

```
master : - Default branch to start with git
         - this branch is created by git for user, any other braches needs to create by user only
         - this branch is also cosidered as official "sourch of truth" code, which is in or will go in production
         - in 2020 github renamed the default branch from "master" to "main", 
           however for git default branch name is still "master"
           
           
HEAD : It is the branch pointer
       this pointer is basically reflect which branch is currenly active for commiting
       
```

```
# list of git branches
git branch  

# Rename Branch
git branch -m <new_name>    : -m is for move

# Creating branch
git branch <branch_name>            : creating new branch having base of "HEAD branch"

# Deleting branch 
First switch to another branch
git branch -D <branch_name>

# switching head to new branch
git switch <existing_branch_name>   
git chekout <existing_branch_name>
Note : Add and commit the changes before switching branches

# reating new_branch and switching towards new_branch
git switch -c <new_branch>          
git chekout -b <new_branch>
```
