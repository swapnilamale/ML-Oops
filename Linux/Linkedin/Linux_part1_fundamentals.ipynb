#### chmod

```
- Linux is multi-user OS
- It works on premise that "Every user has responsibility for controlling access to their files"
- Every file directory created by you can be protected from , made accessible to other users 
  by changing the access permissions

                            # There are 3 types of permissions (rwx)
                            r - read
                            w - write (Read/Modify/delete)
                            x - execute (Running a program)
                                (in case of directory x means, you use cd command for that directory)


                            # These Permissions actually give at 3 levels
                            u - User
                            g - Group (can be people in the same project)
                            o - Other (public : everyone on the system) 
                            a - all


# chmod (change mode) --> chmod (u/g/o/a)(-/+)(r/w/x) <file_name>
chmod g-w <file_name>  : for "group" "remove" "write" permission for "<file_name>"
chmod a-w <file_name>  : for "all"   "remove" "write" permission for "<filename>"
chmod u-w <file_name>  : for "user"  "remove" "write" permission for "<file_name>"
                          In this case even thogh you want to delete the file, you cannont
                          only root account can delete this file now

chmod o+r  <file_name> : for "Other" "Add", "read" permission for "<file_name>"
chmod u+rw <file_name> : for "user" "Add" "read and write" permission for "<file_name>"
chmod +x   <file_name> : "Add" "Executable" Permition for all levels for "<file_name>

chmod u=rwx <file_name>: for "user" give 'rwx' permission for <file_name>



                                        | permission using numbers |
                                                --- : 0
                                                --x : 1
                                                -w- : 2
                                                -wx : 3
                                                r-- : 4
                                                r-x : 5
                                                rw- : 6
                                                rwx : 7

chmod 777 <file_name>    : Give all the permission at all level
chmod -R 777 <directory> : Give all permission at all level to given folder and it all subfolder and all files in it
```

#### chown & chgrp

```
# Change the Ownership of the Permissions
- There are two owners of file & directory
     1. user 
     2. group
  You can change both user & group
 
chown  <user> <file_name>    : change owner to <user> for <file_name>
chgrp -R <group> <file_name> : change group to <grp> for <file_name>

Note : you cannont change root account (Administrator) from user or from group as it owns everything
```

#### ACL (Access Control List)  : getfacl / setfacl

```
Lets have a scenario in which a particular user is not a member of group created by you but still want to give permission to him explicitly without adding him to any group!
he will be consider in other, if we give write permition to other all other user will get write permition
Basically, ACL are used to make flexible permission Mechanism in Linux


getfacl <file_name>    : Displaying Permission of the File

setfacl -m u:<user>:rwx <file_name>    : Add permission to user
setfacl -dm u:<user>:rwx <folder_name> : Add Permission to directory
setfacl -x u:<user> <file_name>        : Remove Permition for perticular user
setfacl -dx u:<user> <file_name>
setfacl -b  <file_name>                : Make permition to default / reset permition

m : modifiy ; x: remove 
u : user

Note: 
1. As you assign ACL permission to file/directory it adds + sign at the end of the permission (ls -l)
2. Setting w permission in ACL: does not allow to remove a file in case of ACL
```

##### Special Permissions 

```
There are 3 additional special permission in linux

# setting a special permition at user level
setuid      :  This permition allows user to run the command where as it will be executed by root in background
chmod u+s <file>   : assign special permission at user level                    (- --s --- --- )
chmod u-s <file>   :  remove special permition at user level

# setting special permition at group level
setgid      :  This permition allows group to run the command where as it will be executed by root in the background
chmod g+s <file>   : assign special permission at group level                   (- --- --s --- )
chmod g-s <file>   :  remove special permition at group level



sticky bit t:  Permiton that allow only user and root to DELETE thoes files (- --- --- --t)
              any user cannot delete that directory
              Ex : /temp --> This is use to protect directory from accidental deletion
chmod +t <file>    : adding sticky bit to the file                              (- --- --- --t )
                     other than user and root , anyone could write the file     
                     will not not be permitted to DELETE the file/directories or subdeirectories



find all the files/commands with special permition (setuid / setgid)
find / -perm /6000 -type f

Note:
Every command is a file in a linux , to find it location we use which
which <command> : location of the command you run
```

#### cut

```
cut
- it is the line utility use to cut parts of lines or piped data & print result to stdout(display)
- it can cut parts of line by delimiter,byte position & character

# Check the version of cut command
cut --version

# Fetching character number & character range from every line
cut -c1 <file>         # first letter from every line
cut -c1,4,7 <file>     # first , forth, seventh letter from every line
cut -c1-5 <file>       # first 5 charecters from every line
cut -c1-3,6-9 <file>   # first to third, sixth to nineth from every line

# Fetching columns by delimeter
cut -d@ -f 1 <file>     # Fetch first column after text seperation by @
cut -d' ' -f 3 <file>   # Fetch third column after text seperation by space

# cut command can be use with other command using pipe operator
ls -l | cut -d' ' -f 1
ls -l | cut -c1-4
```

#### awk

```
awk : This is mostly use to extract the field 
      (column base on delimeter, default delimeter is space)

--> Check the version of awk command
awk --version

--> awk command fetching columns seperated by space
awk '{print}' <file>                   :  print all field seperated by space from <file>
awk '{print $0}' <file>

awk '{print $1}' <file>                :  print first column seperated by space from <file>
awk '{print $1,$3}' <file>             :  print first & third column seperated by space from <file>
awk '{print $NF}' <file>               :  print last field seperated by space from <file>
awk '{print $(NF-1)}' <file>           :  print second last column seperated by space from <file>
awk '{print NF}' <file>                :  print number of columns for each row

awk '{print length($1)}' <file>        :  print first column filed len, where column seperated by space from <file>


--> awk command fetching columns seperated by delimeter
awk -F@ '{print $1}' <file>            : print first column seperated by @ from <file>

--> awk command for searching
awk '/dum_word/ {print}' <file>        :  print all column seperatd by space from <file> contaion a word 'dum_word'

--> awk command for column replacement
awk '{$2="apple";print}' <file>        : print all column seperated by space from <file> 
                                         & replace second filed with "apple"


--> Row Filtering
awk '{NR==2,NR==5 print}' <file>       : print all column seperated by space AND 2nd and 5th row from <file>
awk '$2~/usa|uk/ {print $2}' <file>    : print all column seperated by space AND rows having "usa" or "uk" 
                                         in 2nd column keyword from <file>
awk 'length($0) > 15' <file>           : print all column seperated by space from <file> where first field length > 15
awk '{if($2 == "apple") print}' <file> : print all column seperated by space form <file> where second field is "apple"
```

#### sed : very powerful command

```
- Replacing a string in a file with another string
- find & delte a line
- remove empty lines
- remove the first or n lines in a file
- replace tabs with spaces
- show defined lines from a file
```

```
# Display
sed -n 12,18p <file> : display only line 12 to 18 from <file>
sed 12,18d <file>    : display only line except 12 to 18 from <file>

# Insert
sed G <file>    : put empty line after each line

# Replace a word
sed 's/old_word/new_word/g' <file>      : substitute old_word with new_word
sed '8!s/old_word/new_word/g' <file>    : substitute old_word with new_word except in 8th line
sed 's/old_word//g' <file>              : remove old_word
sed 's/\t/ /g' <file>                   : substitute tab with space


# find base on word & delete line
sed '/<query_word>/d' <file>    : delete the lines contains a word <query_word>
sed '/^$/d' <file>              : delete empty lines
sed '1d' <file>                 : delete first line from the <file>
sed '1,2d' <file>               : delete first two line from the <file>


Note: 
sed command does not change the content in <file> default, 
to reflect changes in <file> we need to use -i : insert changes
Ex : sed -i 's/old_word/new_word/g' <file>   
```

####  zip file (tar / gzip)

```
- "tar" take bunch of file & make it to single container like a single file 
- Note : it does not compress or reduce file size
- After tar we can compress it using "gzip" to make file size smaller

| tar: combining files to one file |
tar cvf <ziped_file_name>.tar <directory>  : ziped given directory
tar cvf <ziped_file_name>.tar .            : ziped current file directory
tar xvf <ziped_file_name>.tar              : Unzip the tar file into current directory

| gzip : compressing & decompressing file |
# Compressing
gzip <ziped_file_name>.tar        
<ziped_file_name>.tar -> compressed file size -> <ziped_file_name>.tar.gz
                                                    (size will be less)
# Decompressing                                                   
gzip -d <ziped_file_name>.tar.gz | gunzip <ziped_file_name>.tar.gz   
<ziped_file_name>.tar.gz --> decompressed file size --> <ziped_file_name>.tar    
                                                    it will take retain its original size
                                                      
```

```
Lets consider an enterprise level scenario where there are 1000 of servers
Now when new users come, for him/her it is very overwhelming to create account in each server.
it become further more complex when you want to update user info or change the password for each server.
In Corperate,Things dont work like that

For that Every Operating sys has "Account Authentication Programm" (Directory Service)
Ex:
Windows has "Active Directory" Program
In this you create an account and your info will be stored in the user-directory-database as user
& whenever you want to login to linux server account authentication goes against the user-directory-database, 
while logging in server will communicate with "Active Directory" using communicator "WinBIND", gives authentication.
Fetch user information and user can log in, in that server

This process of communicating with "Account Authentication Programm" , "Active Directory" in our case
is executed using protocal called LDAP (Lightweight Directory Acess Protocal)


Note : if you want to login to any server, you will need SSH Protocal
       if you want to login to any server, using "Active Directory" , you will need LDAP protocal
       

"Account Authentication Programm" (Directory Service) for different OS
- Windows : Active Directory
- Linux   : IDM (Identity Manager)
            OpenLDAP (Note : LDAP is a protocol not Directory Service)
- IBM Directory Server
- JumpCloud
```
