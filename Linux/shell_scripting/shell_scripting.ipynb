<h2 style="text-align:center; color:blue;">
Variables
</h2>

```
- There should not be any SPACE arround '=' operator while assigning variable
- Use Double quotes (""), while defining a string
- Use "$" (Command of substitution) sign to operate an variable withing double quotes ("")
- while defining command into the variable that command should be surrounded by inverted ticks `<command>` 
          Ex: `hostname`
  Note : These tics are not single inverted comma , it is available right below escape key
```

````python
%%bash
# Defining variables
a=10                                  # Int 
name="Sham"                           # String 
readonly country="INDIA"              # Defining constant Variable (Cannot be change latter or reassign)

 # Accessing a Variable using '$' or '${var}' inside string (fstring)
echo "$a and $name in $country"        # Method 1 --> 10 and Sham in INDIA
echo "${a} and ${name} in ${country}"  # Method 2 --> 10 and Sham in INDIA

# Storing COMMAND OUTPUT TO VARIABLE
DATE_START=$(date)
echo "<date> commnad output   : $DATE_START"

# Storing COMMAND AS VARIABLE
# Method 1: using ``
DATE_NOW_1=`date`   
echo "executing <date> command method 1: $DATE_NOW_1)"

# Method 2: using ""
DATE_NOW_2="date"  
echo "executing <date> command method 2: $($DATE_NOW_2))"


# Removing variable
unset a   
````

<h2 style="text-align:center; color:blue;">
Arrays
</h2>

```
- values should be define space seperated
- Array can store heterogenious datatype
- Indexes are started from 0
```

```
                        positive_index     0  1     2        3
                                         103 2.3 "Hello" "Chetan"
                        negative index    -4  -3   -2       -1
```

````python
%%bash
# Defining an Array
name="Chetan"
arr=(103 2.3 "Hello" $name)   # Elements should be space seperated

# len of array
echo ${#arr[*]}    # 4 

# Accessing all the values in the array
echo ${arr[@]}    # 1 2.3 Hello Chetan
echo ${arr[*]}  


# Acessing First element in the array
echo $arr        # 1
echo ${arr[0]}

# Accessing last element in the array
echo ${arr[-1]} # Chetan

#Accesing element
echo ${arr[3]}      # Value at 3rd Index (Chetan)

#Slicing
echo ${arr[*]:0:2}  # start with the 0 index and go 1 step ahead    # 1 2.3
echo ${arr[*]:2}    # get all values from second index              # Hello Chetan
echo ${arr[*]:1:3}  # start with the 1 index and go 2 step ahead    # 2.3 Hello Chetan

# Updating array with the new values
arr+=("New" 30 40)        # Appending Elements in the array
echo ${arr[*]}
````

<h2 style="text-align:center; color:blue;">
Arithmatics
</h2>

```
                                                +  : Addition
                                                -  : Substraction
                                                *  : Multiplication
                                                /  : Division
                                                %  : Remender
                                                ** : Power

- Arithmatic Operator can be used using ((__)) & $((...)) or with let command

 ((...))	Arithmetic evaluation → executes an arithmetic expression directly (no $), 
                                    used for updating variables.
 $((...))	Arithmetic expansion → produces a value you can assign or print.
```

````python
%%bash
echo "((10+3))"
echo "$((10+3))"
````

````python
%%bash
a=3
b=10

# Note $ below is use for printing purpose with echo
echo "SUM : $((a+b))"
echo "Subtraction : $((a-b))"
echo "Multiplication : $((a*b))"
echo "Floor Division : $((a/b))"
echo "Reminder : $((b%a))"
echo "Power : $((a**b))"
````

````python
%%bash
# let command
a=3
b=10

let SUM=$a+$b
let Subtraction=$a-$b
let Multiplication=$a*$b
let Floor_Division=$a/$b
let Reminder=$b%$a
let Power=$a**$b

# Note $ below is use for printing purpose with echo
echo $SUM
echo $Subtraction
echo $Multiplication
echo $Floor_Division
echo $Reminder
echo $Power
````

````python
%%bash
# Increment or Decrement value of variable
a=5
echo "$a"      # prints 5

a=$((a+1))     # Method 1: using arithmetic expansion (Note here $ is use for assigning arithmatic expansion 
               #                                            not for printing)
echo "$a"        # prints 6

((a++))        # Method 2: using arithmetic evaluation
echo "$a"        # prints 7

let a++        # Method 3 : using let
echo $a        # prints 8

((a+=3))       # Method 2: usin"g arithmetic evaluation
echo "$a"        # prints 7

let a+=3
echo "$a"
````

<h1 style="text-align:center; color:red;">
if-then-elif-then-else-fi  
</h1>

```
- condition is defined in square brackets : [ condition ] 
- instead of single bracket we can use double square bracket as well which is functionaly more rich :  [[ condition ]]
- there should be SPACE at start and end of square braket  AND before and after of comparison operator 
  [ __ comparison_operator __ ]
- logical operatory can be used between two square brackets [ condition1 ] logical_operator [ condition2 ]
```

````python
%%bash
marks=85
if [ $marks -gt 80 ]        
then 
    echo "First Division"
elif [ $marks -ge 60 ]
then 
    echo "Second Division"
else
    echo "Your are Failed"
fi
````

````python
%%bash
age=30
area="INDIA"
if [[ $age -gt 18 ]] && [[ $area == "INDIA" ]]
then 
    echo "You can vote"
else
    echo " You Cant Vote"
fi
````

````python
%%bash
# Note handling a string variable with mulitple words seperated by space,
# in if statement condition when using single square brackets [], else use [[ ]]
a="Hello there"
if [ "$a" == "Hello there" ]  #" Note "$a" should be in "" only, it will reflect passing multiple argument error
then                          
    echo $a
fi
````


````python
%%bash
count=0
num=5

while [ $count -le $num ]
do 
    echo $count
    sleep 0.5
    let count++
done
````

````python
%%bash
num=5
while [ $num -gt 0 ]
do
    echo "$num secs remaining to stop the process!"
    sleep 1
    let num--
done
echo "||||| BANG |||||"
````

<h2 style="text-align:center; color:blue;">
Break and Continue in loop
</h2>

```
break : stop the loop
```

````python
%%bash
# WAS to confirm if a given no is present in array or not
query_num=6
arr=(1 5 2 7 5 6 23 7 5 6)

for i in ${arr[@]}
do
    if [ $query_num -eq $i ]
    then
        echo "$query_num is found !!!"
        break 
    fi
    echo $i
done
````

```
continue : stop current interation of the loop and start next iteration
```

````python
%%bash
# WAS to print only odd values in array
arr=(1 5 2 7 5 6 23 7 5 6)
for i in "${arr[@]}"
do 
    let r=$i%2
    if [ $r -eq 0 ]
    then    
        continue 
    fi
    echo $i
done
````

<h1 style="text-align:center; color:red;">
FUNCTION
</h1>

```
- Input arguments inside function is accessed by $1,$2,$3
- local variable can be define inside function 
  local <variable>=<value>
```

````python
%%bash
# -------------  Function ----------------#
# Defining a function 
my_function(){
    # Taking Input in Function Defination
    local num1=$1   # First Argument
    local num2=$2   # Second Argument
    global_var=256  # global variable, will be valid outside of the function as well

    local sum=$((num1+num2))
    echo $sum  # returned value
    }

# Calling a function and sending argument to the function
my_function 2 4

# storing returned value
result=$(my_function 2 -4)
echo "Return Value is : $result"                 

# global variable outside the function
echo $global_var
````

````python
%%bash
# WAS to check the divsibility of input number by 1 to 10
divisibility(){
    local num=$1
    for i in {1..10}
    do
        local rem=$(( $num%i ))
        if [ $rem -eq 0 ]
        then
            echo "$num is divisible by $i"
        fi
    done
}

divisibility 66
````
